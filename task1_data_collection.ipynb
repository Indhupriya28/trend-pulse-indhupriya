```python
import requests


try:
  url="https://hacker-news.firebaseio.com/v0/topstories.json"
  id = requests.get(url).json()
  id_json = id[:500]
  print(id_json)
except requests.exceptions.HTTPError:
  print("Failed processing the request")

```

by using requests object, api call is triggered. Response will be list of ids. Here id_json has first 500 data from the fetched result.

```python
import requests
import time
from datetime import datetime

headers = {"User-Agent": "TrendPulse/1.0"}
json_data=[]
try:
  limit =25

  cat_count = {
      "technology" : 0,
      "worldNews" :0,
      "sports" :0,
      "science":0,
      "entertainment":0
  }

  collected_story = []

  technology = ["AI", "software", "tech", "code", "computer", "data", "cloud", "API", "GPU", "LLM"]
  worldNews = ["war", "government", "country", "president", "election", "climate", "attack", "global"]
  sports = ["NFL", "NBA", "FIFA", "sport", "game", "team", "player", "league", "championship"]
  science = [	"research", "study", "space", "physics", "biology", "discovery", "NASA", "genome"]
  entertainment = ["movie", "film", "music", "Netflix", "game", "book", "show", "award", "streaming"]


  for id in id_json:
    res = requests.get(f"https://hacker-news.firebaseio.com/v0/item/{id}.json",headers=headers).json()

    title = res.get("title").lower()

    if cat_count["technology"] < limit and any( word.lower() in title for word in technology):
        category= "technology"
        time.sleep(2)

    elif cat_count["worldNews"] <limit and any(word.lower() in title for word in worldNews):
        category = "worldNews"
        time.sleep(2)

    elif cat_count["sports"] <limit and any(word.lower() in title for word in sports):
        category= "sports"
        time.sleep(2)

    elif cat_count["science"] <limit and any(word.lower() in title for word in science):
        category = "science"
        time.sleep(2)

    elif cat_count["entertainment"] <limit and any(word.lower() in title for word in entertainment):
        category = "entertainment"
        time.sleep(2)

    cat_count[category] +=1
    res["category"] = category

    if category:
        extracted_story = {
            "post_id": res.get("id"),
            "title": res.get("title"),
            "category": category,
            "score": res.get("score", 0),
            "num_comments": res.get("descendants", 0),
            "author": res.get("by"),
            "collected_at": datetime.now().isoformat()
        }

    collected_story.append(extracted_story)

  print("Collected:", len(collected_story))
  print(collected_story[0:100])
except requests.exceptions.HTTPError:
  print("Failed processing the request")

```

1.Since the limit is provided as 125,  splitted the each categoty limit by 25. Iterating the list of ID, api request triggered to get the object data and categorising it based on the keywords and also the size of each category.
2.Maintaining the dictionary to count the each category item.TOmake sure length doesn't exceed limit 25 (per category).

3.To display the listed 7 fields for each story, created a list. For each iteration, processed story is apppended to the collected_story list (7 fields)

```python
from datetime import datetime
import os
import json

os.makedirs(f"data",exist_ok=True)
filename = f"data/trendpulse_{datetime.now().strftime('%Y%m%d')}.json"

with open(filename, "w", encoding="utf-8") as f:
    json.dump(collected_story, f, indent=4)

print(f"collected {len(collected_story)} stories.Saved to {filename}")
```
