```python
import pandas as pd

#read the csv files from the google colab
df = pd.read_csv("/data/trends_clean.csv")


#display the rows and columns
print(f"Loaded data: {df.shape}")
#display first 5 data
print(f"First 5 rows:")
print(df.head())
#diplay the avg
print(f"\nAverage score : {df["score"].mean()}")

print(f"Average comments: {df["num_comments"].mean()}")
```

```python
import numpy as np

score = df["score"].values
mean_score = np.mean(score)
median_score = np.median(score)
std_score = np.std(score)

print("--- NumPy Stats ---")
print(f"Mean score: {mean_score}") #average of the score
print(f"Median score: {median_score}")#middle value of the score
print(f"Standard deviation:{std_score}") #standard deviation

print(f"Highest score: {np.max(score)}") #highest score
print(f"Lowest score: {np.min(score)}") #lowest score

```

```python
import pandas as pd


category = df["category"].values

#print(np.unique(category,return_counts=True))
#Return the list of unique category and list of the their counts
uniq_category, count = np.unique(category, return_counts=True)

#fetch the index of the highest count
max_index = np.argmax(count)
print(f"Most stories in : {uniq_category[max]} ({count[max_index]} stories)")

#fetch all the num_comments values
comment = df["num_comments"].values
#get the hightest comments index value
most_com_index = np.argmax(comment)

#get the story of the index value
story = df.iloc[most_com_index]

print(f"Most commented story: {story["title"]} - {story["num_comments"]} comments ")
```

```python
#Adding engagement and is_popular columns into existing dataframe
df["engagement"] = df["num_comments"] / df["score"] +1
df["is_popular"] = df["score"] > df["score"].mean()
print(df.head())

```

```python
path = "/data/trends_analysed.csv"
df.to_csv(path, index=False)
print(f"Saved to {path}")
#saving the updated dataframe to the trends_analysed csv file
```
