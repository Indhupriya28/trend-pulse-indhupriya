```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/trends_analysed.csv")
top_stories = df.sort_values(by="score", ascending=False).head(10)
#it get the top 10 stories.by sorting it in descending order
```

```python
short_titles=[]

#This reduce the length of the title. Slicing the title
for tit in top_stories["title"]:
  if len(tit) >50:
    short_title = tit[:50]
  else:
    short_title = tit
  short_titles.append(short_title)

top_stories["short_title"] = short_titles

```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.barh(top_stories["short_title"],top_stories["score"])
plt.title("Top 10 stories ")
plt.xlabel("short_title")
plt.ylabel("score label")

plt.savefig("outputs/chart1_top_stories.png")
plt.show()

color = ["blue","green","yellow","red","purple"]
category_count = df["category"].value_counts()
plt.bar(category_count.index, category_count.values,color= color)
plt.title("Number of stories per category")
plt.xlabel("Category")
plt.ylabel("Number of stories")
plt.savefig("outputs/chart2_top_stories.png")
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.scatterplot(data = df,x="score",y="num_comments",hue="is_popular")

plt.title("Score Vs Number of comments")
plt.xlabel("Score")
plt.ylabel("Number of Comments")
plt.savefig("outputs/chart3_top_stories.png")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].barh(top_stories["short_title"],top_stories["score"])
axes[0].set_title("Top 10 stories ")
axes[0].set_xlabel("short_title")
axes[0].set_ylabel("score label")



color = ["blue","green","yellow","red","purple"]
category_count = df["category"].value_counts()
axes[1].bar(category_count.index, category_count.values,color= color)
axes[1].set_title("Number of stories per category")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Number of stories")


axes[2].scatter(data = df,x="score",y="num_comments",label="is_popular")

axes[2].set_title("Score Vs Number of comments")
axes[2].set_xlabel("Score")
axes[2].set_ylabel("Number of Comments")

fig.suptitle("TrendPulse Dashboard", fontsize=16)

plt.tight_layout()
plt.savefig("outputs/dashboard.png")
plt.show()
```
