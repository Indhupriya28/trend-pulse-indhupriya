```python
import pandas as pd
import json

#read the json file using pandas
json_df = pd.read_json("/data/trendpulse_20260410.json")
print(len(json_df))
print(json_df.describe())
print(json_df.info())
```

```python
clean_df = json_df.copy()

print(clean_df["post_id"].duplicated().sum())
print(clean_df.isnull().sum())
#display the count of duplicated and null values
```

```python
#REmove the duplicates value in the post_id
clean_df = clean_df.drop_duplicates(subset="post_id")
print(f"After removing duplicates : {len(clean_df)}")

#drop null values if any present in post_id, title, score
clean_df = clean_df.dropna(subset=["post_id", "title", "score"])
print(f"After removing nulls : {len(clean_df)}")

#set the score and num_comments type as integer
clean_df["score"] = clean_df["score"].astype(int)
clean_df["num_comments"] = clean_df["num_comments"].astype(int)

#this remove the score value < 5
clean_df[clean_df["score"] >=5]
print(f"After removing low scores : {len(clean_df)}")

#remove the extra space iin the title
clean_df["title"]= clean_df["title"].str.strip()

print(f" Loaded {len(clean_df)} stories from trendpulse_20260409.json")

```

```python
#write the cleaned data into csv file
path = "/data/trends_clean.csv"
clean_df.to_csv(path, index=False)

print(f"\nCleaned data saved to {path}")
print("Rows saved:", len(clean_df))
#display the category with the count value
print(f"\n Stories per category:\n")
print(clean_df["category"].value_counts())
```
